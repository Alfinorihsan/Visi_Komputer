Implementasi Face Detection

## 1. Import Library:
Pertama, kita perlu mengimpor library yang diperlukan. Kita akan menggunakan OpenCV untuk melakukan face detection dan Matplotlib untuk menampilkan gambar.

```python
import cv2
from matplotlib import pyplot as plt
```

## 2. Load Cascade Classifier:

Setelah mengimpor library, kita perlu memuat pre-trained Haar Cascade Classifier yang akan digunakan untuk mendeteksi wajah dalam gambar.

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

## 3. Load Image:
Selanjutnya, kita memuat gambar yang ingin kita deteksi wajahnya.

```python
image = cv2.imread('gambar.jpg')
```

## 4. Convert to Grayscale:
Agar lebih mudah untuk mendeteksi wajah, kita akan mengonversi gambar ke mode grayscale.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## 5. Detect Faces:
Kita menggunakan metode detectMultiScale() dari face_cascade untuk mendeteksi wajah dalam gambar.

```python
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

## 6. Draw Rectangles Around Faces:
Setelah wajah dideteksi, kita akan menggambar persegi panjang di sekitar setiap wajah yang terdeteksi.

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

## 7. Display Image with Detected Faces:
Terakhir, kita akan menampilkan gambar dengan wajah yang terdeteksi.

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
